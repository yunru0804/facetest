```python
#百度下載圖片
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from time import sleep
import json
import urllib
import os
import pandas as pd

# 啟動瀏覽器工具的選項
options = webdriver.ChromeOptions()
# options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--incognito")               #開啟無痕模式
options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

driver = webdriver.Chrome( options = options )

photolimit=19
listLink=[]


#取出要使用的名單
csv_data=pd.read_csv("中國女演員.csv")
csv_df=pd.DataFrame(csv_data)
df_ch=csv_df['Name'].values.tolist()
#df_ch=csv_df['Name'].values.tolist()[452:]   #若連線斷掉，可先確認下載到哪個名單，讓名單接著跑


def visit():
    # 前往指定連結
    driver.get('https://image.baidu.com/search/index?tn=baiduimage&ipn=r&ct=201326592&cl=2&lm=-1&st=-1&fm=detail&fr=&hs=0&xthttps=111110&sf=1&fmq=1637048415250_R&pv=&ic=0&nc=1&z=&se=&showtab=0&fb=0&width=&height=&face=0&istype=2&ie=utf-8&word=%E9%9D%A2%E7%9B%B8&oq=%E9%9D%A2%E7%9B%B8&rsp=-1')
    
def search():
    driver.find_element(By.CSS_SELECTOR, "input#kw").clear()
    for df in df_ch:
        driver.find_element(By.CSS_SELECTOR, "input#kw").send_keys(df)
        sleep(3)
        driver.find_element(By.CSS_SELECTOR, "input.s_btn").click()
        sleep(3)
        getUrl(df)
        sleep(3)
        driver.find_element(By.CSS_SELECTOR, "input#kw").clear()
        sleep(3)
    
# 取得每個項目的 url
def getUrl(df):
    # 取得主要元素的集合
    a_elms = driver.find_elements(By.CSS_SELECTOR, 'div.imgpage li.imgitem div.imgbox-border img')
    
    # 逐一檢視元素
    for index, a in enumerate(a_elms):
        if index > photolimit:
            break
        aLink = a.get_attribute('src')
        
        # 放資料到 list 中
        listLink.append({
            'name':df+"_"+str(index+1),
            'link':aLink})        
    
    saveJson()
    
def saveJson():
    fp = open("actress_china.json", "w", encoding='utf-8')
    fp.write( json.dumps(listLink, ensure_ascii=False) )
    fp.close()
    
# 關閉瀏覽器
def close():
    driver.quit()

def download():
    fp = open("actress_china.json", "r", encoding='utf-8')
    strJson = fp.read()
    fp.close()
    
    folderPath = 'actress_china'
    if not os.path.exists(folderPath):
        os.makedirs(folderPath)
    
#     將 json 轉成 list (裡面是 dict 集合)
    listResult = json.loads(strJson)

    #下載圖片
#     for obj in listResult[8370:]:   #若圖片當掉，可去資料夾確認下載筆數，再設定繼續run
    for obj in listResult:
        try:
            urllib.request.urlretrieve(obj['link'], os.path.join(folderPath , obj['name']+'.jpg'))
            print(obj['name']+'.jpg 已下載')
        except urllib.error.HTTPError as exception:
            print(obj['name']+".jpg 無法下載")
            continue
        except TimeoutException as e:
            continue
        

if __name__ == '__main__':
   # visit()
  #  search()
 #   close()
   download()
```
