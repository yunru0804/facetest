```python
#Google下載圖片(測試中)
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import json,csv
import urllib
import os,re
import pandas as pd

# 啟動瀏覽器工具的選項
options = webdriver.ChromeOptions()
# options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--incognito")               #開啟無痕模式
options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

# # 使用 Chrome 的 WebDriver (含 options)
driver = webdriver.Chrome( options = options )

photolimit=10

#若卡住，尚未輸入CSV的話，需註解再執行，不然前面跑的都會歸零
#listLinks=[]


#取出要使用的名單
csv_data=pd.read_csv("kr_singer.csv")
#df_ch=csv_data['Name'][201:]
df_ch=csv_data['Name'][1054:]
def visit():
    # 前往指定連結
    driver.get('https://tw.images.search.yahoo.com/search/images;_ylt=AwrtFIqSnaNhp0MAKqFt1gt.?p=%E5%91%A8%E6%9D%B0%E5%80%AB&ei=UTF-8&fr=sfp&imgf=portrait&fr2=p%3As%2Cv%3Ai')
    
def search():
    driver.find_element(By.CSS_SELECTOR, "input.yschsp").clear()
    for df in df_ch:
        try:
            driver.find_element(By.CSS_SELECTOR, "input.yschsp").send_keys(df)
            sleep(2)
            driver.find_element(By.CSS_SELECTOR, "input.ygbt").click()
            sleep(2)
            getUrl(df)
            sleep(2)
            driver.find_element(By.CSS_SELECTOR, "input.yschsp").clear()
            sleep(2)
        except Exception as e:
            print(df+"  搜尋不到圖片")
            driver.find_element(By.CSS_SELECTOR, "input.yschsp").clear()
            continue

def getUrl(df):
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a.img"))).click()
    sleep(2)
    for i in range(photolimit):
        a_elms=driver.find_elements(By.CSS_SELECTOR, 'div.cont')
        a11_regex=r'\[\'(.+)\'\]'
        for a in a_elms:
            imgs=a.find_elements(By.CSS_SELECTOR,'img#img')
            a11 = [j.get_attribute('src') for j in imgs]
            a111=str(a11)
            title=a.find_elements(By.CSS_SELECTOR,'.title')
            a21 = [j.get_attribute('innerText') for j in title]
            listLinks.append({
                'name':df+"_"+str(i+1),
                'link':re.match(a11_regex,a111).group(1),
                'title':a21
            }) 
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[title='下一張圖片']"))).click()
            sleep(2)
            
def saveCsvs():
    with open('kr_singer1.csv', 'a', newline='', encoding='utf-8') as f:
        for index, i in enumerate(listLinks):
            r = csv.DictWriter(f, fieldnames=i.keys())
            if index == 0:
                r.writeheader()
                r.writerow(i)
            else:
                r.writerow(i)
    
# 關閉瀏覽器
def close():
    driver.quit()

def download(): 
    folderPath = 'kr_singer'
    if not os.path.exists(folderPath):
        os.makedirs(folderPath)
    
    listResult=pd.read_csv('kr_singer1.csv')
    aa=listResult[1904:]

#     for i in range(len(listResult)):       
    for i in range(len(aa)):
        try:
#             urllib.request.urlretrieve(listResult['link'][i], os.path.join(folderPath , listResult['name'][i]+'.jpg'))
            urllib.request.urlretrieve(listResult['link'][i+1904], os.path.join(folderPath , listResult['name'][i+1904]+'.jpg'))

#             print(listResult['name'][i]+'.jpg 已下載')
            print(listResult['name'][i+1904]+'.jpg 已下載')

        except Exception as e:
            continue
    print("done")

if __name__ == '__main__':
#     visit()
#     search()
#     saveCsvs()
#     close()
   download()
```
